Fit a 10 -nearest neighbors model to predict the Sale Price of a home in each neighborhood of Ames. That is, the categorical variable Neighborhood should be the only input feature to your model.

A DataFrame consisting of this "test data" is provided for you.

```python
from sklearn.preprocessing import OneHotEncoder

# neighborhood is categorical
X = df[["Neighborhood"]]
y = df["SalePrice"]

# Split into training and validation
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2
)

# needed for categorigal values
encoder = OneHotEncoder( handle_unknown="ignore")

X_train_encoded = encoder.fit_transform(X_train)

X_val_encoded = encoder.transform(X_val)

# train the model
model = KNeighborsRegressor(n_neighbors=10)
model.fit(X_train_encoded, y_train)

# from the exercise
X_test = pd.DataFrame({
    "Neighborhood": df["Neighborhood"].unique()
})

X_test_encoded = encoder.transform(X_test)

predictions = model.predict(X_test_encoded)

print("\nExercise 2:")
for neigh, pred in zip(X_test["Neighborhood"], predictions):
    print(f"{neigh}: ${pred:,.2f}")
```

### How it works is it takes a point and matches it to the closest 10 neighbors. then it averages their prices.
